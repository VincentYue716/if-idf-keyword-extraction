```python
import pandas as pd

def read_csv_with_weight(csv_path):
    df = pd.read_csv(csv_path, encoding='utf-8', low_memory=False)
    # 转发数,评论数,点赞数
    comments = df['评论数'].tolist()
    reposts = df['转发数'].tolist()
    likes = df['点赞数'].tolist()
    weibos = df['微博正文'].tolist()
    results = []

    # for row in range(len(weibos)):
    #     if comments[row] + reposts[row] + likes[row] > 10:
    #         results.append(weibos[row])


    for row in range(len(weibos)):
        times = 1 + int(reposts[row]/100) + int(comments[row]/500) + int(likes[row]/1000)
        for time in range(times):
            results.append(weibos[row])

    new_list = []
    for result in results:
        result = result.split()
        new_list.append(result)

    return new_list
```

```python
from gensim.corpora import Dictionary
from gensim.models import LdaMulticore

trump1 = read_csv_with_weight('./trump/trump_1.csv')
trump2 = read_csv_with_weight('./trump/trump_2.csv')
trump3 = read_csv_with_weight('./trump/trump_3.csv')
trump4 = read_csv_with_weight('./trump/trump_4.csv')
trump5 = read_csv_with_weight('./trump/trump_5.csv')
trump6 = read_csv_with_weight('./trump/trump_6.csv')
trump7 = read_csv_with_weight('./trump/trump_7.csv')
trump8 = read_csv_with_weight('./trump/trump_8.csv')
trump9 = read_csv_with_weight('./trump/trump_9.csv')
trump10 = read_csv_with_weight('./trump/trump_10.csv')
```

```python
print(len(trump1))
print(len(trump2))
print(len(trump3))
print(len(trump4))
print(len(trump5))
print(len(trump6))
print(len(trump7))
print(len(trump8))
print(len(trump9))
print(len(trump10))
print(len(trump1)+len(trump2)+len(trump3)
      +len(trump4)+len(trump5)+len(trump6)
      +len(trump7)+len(trump8)+len(trump9)+len(trump10))
```

```python
dictionary = Dictionary(trump1)
dictionary.filter_extremes(no_below=5, no_above=0.85, keep_n=10000)
bow_corpus = [dictionary.doc2bow(doc) for doc in trump1]

lda_model = LdaMulticore(bow_corpus,
                           num_topics = 30,
                           id2word = dictionary,
                           passes = 10,
                           workers = 4)

item = []
for j in range(len(trump1)):
    item = item + trump1[j]
# unseen_document = item.split()

bow_vector = dictionary.doc2bow(item)

print("1月")
for index, score in sorted(lda_model[bow_vector], key=lambda tup: -1*tup[1]):
    print("Score: {}\t Topic: {}".format(score, lda_model.print_topic(index, 15)))
```
